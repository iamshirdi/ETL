```
f3=os.listdir("/content/train/Baby Clothes")
f4=os.listdir("/content/valid/Baby Clothes")
f5=os.listdir("/content/train/Random Images")
f6=os.listdir("/content/valid/Random Images")


n=75

for c1,i1 in enumerate(f1):
  try:
    a1=open_image('/content/downloads/Baby Clothes/'+i1)
  except:
    continue
  #train
  if c1<n:
    if i1 not in f3:copyfile("downloads/Baby Clothes/"+i1,"/content/train/Baby Clothes/"+i1 )
  else:
    if i1 not in f4:copyfile("downloads/Baby Clothes/"+i1,"/content/valid/Baby Clothes/" +i1)
      
   #valid
for c2,i2 in enumerate(f2):
  try:
    a1=open_image('/content/downloads/Random Images/'+i2)
  except:
    continue
  if c2<n:
    if i2 not in f5:copyfile("downloads/Random Images/"+i2,"/content/train/Random Images/"+i2 )
  else:
    if i2 not in f6:copyfile("downloads/Random Images/"+i2,"/content/valid/Random Images/"+i2 )
```

```
tfms = get_transforms(do_flip=False)
#path = '/content/drive/My Drive/Colab Notebooks/fastai/'
path = '/content/'
data = ImageDataBunch.from_folder(path,ds_tfms=tfms, size=100)
```

```
data.show_batch(rows=3, figsize=(10,10))
```

```
#!ls '/content/drive/My Drive/Colab Notebooks/fastai/'
```

```
my_trained_mod = create_cnn(data, models.resnet34, metrics=error_rate)
my_trained_mod.fit_one_cycle(6)
```
