```python
import boto3
s3 = boto3.client('s3')
```

```python
resp=s3.list_objects_v2(Bucket='websitename.s3bucket')
keys = []
for obj in resp['Contents']:
    keys.append(obj['Key'])
```

```python
print(len(keys))
```

```python
a=keys[2]
a
```

```python
url='http://websitename.s3bucket.s3-website-us-west-1.amazonaws.com/'
```

```python
from PIL import Image
import requests
response=requests.get(url+a)
from io import BytesIO
img = Image.open(BytesIO(response.content))
img
```
