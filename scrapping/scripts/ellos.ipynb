```python
%%capture cap

import json
import requests
import time
from random import randint
from bs4 import BeautifulSoup
import urllib.request
import csv

gender='girl'
for cg in cg_list:
    category=cg
    main_url='https://www.ellos.se/barn/flicka-stl-86-140/'
    sub_url=main_url+cg
    html_doc=requests.get(sub_url).content.decode()
    soup = BeautifulSoup(html_doc, 'html.parser')
    try:
        pages=soup.find("ul",attrs={"class":"gridwall-pagination"})
        pages_no=len(soup.find_all("li",attrs={"class":"hidden-xs page-item"}))
        num=int(pages_no)+2
        page_urls=[]
        for p_url in range(1,num):
            page_urls.append(sub_url+'?sort=relevance&page='+str(p_url))
    except:
        page_url=sub_url
        
    for page_url in page_urls:
        print(page_url)
        html_doc2=requests.get(page_url).content.decode()
        soup2 = BeautifulSoup(html_doc2, 'html.parser')
        ps=soup2.find_all("div",attrs={"class":"card-image-wrap js-card-image-wrap"})
        print(len(ps))
        for products in ps:
            product=products.find("a",attrs={"class":"js-open-product-layer card-image-link js-card-image-link"})
            url='https://www.ellos.se'+product['href']
            print(url)
            html_doc3=requests.get(url).content.decode()
            soup3 = BeautifulSoup(html_doc3, 'html.parser')
            brand=soup3.find("a", attrs={"class":"product-desc-collection"} ).text
            name=soup3.find("h1",attrs={"class":"product-info-name"}).text
            img_url='https:'+soup3.find("div",attrs={"class":"product-main-photo js-product-main-photo"}).find('a')['href']
            price=soup3.find("span",attrs={"class":"price-amount"}).text
            article_no=soup3.find("span",attrs={"class":"product-sku"}).text
            ac='ellos'+'/'+article_no+str('.jpg')
        #             opener = urllib.request.URLopener()
        #             opener.addheader('User-Agent', 'whatever')
        #             filename, headers2 = opener.retrieve(img_url, ac)

            rr = requests.get(img_url, allow_redirects=True)
            open(ac, 'wb').write(rr.content)

            driver= webdriver.Chrome('chromedriver',options=options)
            driver.get(url)
            driver.set_window_size(1920,1080)
            elem=driver.find_element_by_tag_name('select')
            driver.execute_script("$(arguments[0]).click();", elem)
            sz=driver.find_elements_by_xpath('//span[@class="variantSizeText"]')
            for sze in sz:
                size=sze.get_attribute('innerHTML')
                with open("ellos.csv", "a", encoding='utf-8') as f:                   
                    writer = csv.writer(f)
                    writer.writerow([name,brand,gender,category,size,price,url,img_url,article_no])
            driver.quit()
            with open('log.txt', 'a+') as f:
                f.write(cap.stdout)
                f.write(cap.stderr)
```
