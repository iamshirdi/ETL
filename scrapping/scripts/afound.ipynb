```python
import requests
from bs4 import BeautifulSoup
gender='unisex'


for count,sub_url in enumerate(urls):
    
    category=categorys[count]
    gender_url=sub_url+'#?p_gender=U'
    
    driver= webdriver.Chrome('chromedriver',options=options)
    driver.get('https://www.afound.com/sv-se/barn/babyklader-0-24man/sparkdrakter-bodys')
    driver.set_window_size(1020,720)
    pages=driver.find_elements_by_css_selector('span.paginator__page')

    page_ur=pages[-1].get_attribute("innerHTML")
    page_urls=[]

    for pg in page_ur:
        page_urls.append(gender_url+'#?page='+pg)
    driver.quit()
    print(page_urls)

    for pg_url in page_urls:
        print(pg_url)
        html_doc=requests.get(pg_url).content.decode()
        soup = BeautifulSoup(html_doc, 'html.parser')
        products=soup.find_all("div", attrs={"class":"nproductcard"})
        for pr in products:
            if pr.a['href']=='':
                print('None value breaking')
                break
            try:
                url=pr.a['href']
                html_doc2=requests.get(url).content.decode()
                soup2 = BeautifulSoup(html_doc2, 'html.parser')
                brand=soup2.find("h1", attrs={"class":"product-page__meta-brand"}).a.text
                name=soup2.find("span", attrs={"class":"product-page__meta-brand__product-name"}).text
                article_no=soup2.find("div", attrs={"class":"product-page page page--wide-content"})['data-gtm-id']
                img_url='https://www.afound.com'+soup2.find("div", attrs={"class":"product-attribute-selector__values carousel"}).find("img")['src']

                ac='afound'+'/'+article_no+str('.jpg')
            #             opener = urllib.request.URLopener()
            #             opener.addheader('User-Agent', 'whatever')
            #             filename, headers2 = opener.retrieve(img_url, ac)

                rr = requests.get(img_url, allow_redirects=True)
                open(ac, 'wb').write(rr.content)



                price=soup2.find("h2", attrs={"class":"product-page__meta-cost"}).span.text
                info=soup2.find("div", attrs={"class":"product-page__actions-item"})
                print(name,brand,price,url,img_url)
                driver= webdriver.Chrome('chromedriver',options=options)
                driver.get(url)
                driver.set_window_size(1020,720)
                szs=driver.find_elements_by_css_selector('option')
                for sz in szs:
                    size=sz.text
                    with open("afound.csv", "a", encoding='utf-8') as f:
                                    writer = csv.writer(f)
                                    writer.writerow([name,brand,gender,category,size,price,url,img_url,article_no])

                driver.quit()
            except:
                continue
        
```
