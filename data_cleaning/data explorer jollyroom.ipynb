```python
import pandas as pd
pd.set_option('display.max_colwidth', -1) #full colunm dispaly

result=pd.read_csv("Image-Classification-and-Price-prediction/scrapping/data/jollyroom.csv", encoding='utf-8',index_col=False)
print(result.count())
result.head(3)
```

```python
df=result.drop_duplicates()
result.drop_duplicates().count()
```

```python
df['price'].unique()
```

```python
df2=df
df2['price']=df2['price'].str.strip()
df2['price']=df2['price'].str.strip(' kr')
df2['price']=df2['price'].str.replace(' ','')

df2['price'].unique()
```

```python
pd.options.display.float_format = '{:,.0f}'.format
df2.price=df2.price.str.replace(",", ".")
df2['price']=df2['price'].astype(float)
df2['price']=df2['price'].round()
df2['price']=df2['price'].astype(int)

df2.head(2)
```

```python
df2['size'].unique()
```

```python
sz_list=df2['size'].tolist()

new_sz=[]
for s in sz_list:
    if s.find(" år")!=-1:
        s=s.replace(" år","Y")
        
    if s.find(" cm")!=-1:
        s=s.strip(" cm")
    if s.find(" m")!=-1:
        s=s.split(" m")[0]+"M"
    if s.find(" kg")!=-1:
        s=s.replace(" kg","M")
        
    if s.find("/")!=-1:
        s=s.split("/")[1]
        
    if s.find("Stl ")!=-1:
        s=s.replace("Stl ","")
        if len(s)>2:
            s='Na'
    if s in [ 'Baby', 'Junior',]:
        s='Na'
    if s in ['S-M','L-XL','M-L','XS-S',]:
        s=s.split('-')[1]
    if s.find("onesize")!=-1:
        s='One Size'
#     if s in ['12-18 kg','3-6 kg']:
#         s='Na'
    
    new_sz.append(s)
set(new_sz)
```

```python
df2['new_size']=new_sz
```

```python
colunms=['article_no','gender','category','price','new_size','brand']
df3=df2[colunms]
df3.head(2)
```

```python
dfr=df3
dfr.to_csv('jollyroom_cleaned.csv',index=False) #for rows not showing
```
