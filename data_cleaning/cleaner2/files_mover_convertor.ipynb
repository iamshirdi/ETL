```python
raw_dir="scrapping/data/"
import os
f1=os.listdir(raw_dir)
f2=[]
for f in f1:
    if f.endswith('.csv'):
        f2.append(f)
print(len(f2))
f2
```

```python
# Creating an empty Dataframe with column names only
import pandas as pd
dfObj = pd.DataFrame(columns=['article_no','gender','category','price','size','brand','url','name'])
print("Empty Dataframe ", dfObj, sep='\n')
```

```python
for fi in f2:
    name=fi.strip('.csv')
    df=pd.read_csv(raw_dir+fi, encoding='utf-8',index_col=False)
    colunms=['article_no','gender','category','price','size','brand','url','name']
    df2=df[colunms]
    df2['article_no']=name+'/'+df2['article_no'].astype(str)
    break
df2.head(2)
```
