```python
pd.options.display.float_format = '{:,.0f}'.format
df2=df
df2.price=df2.price.str.strip(' kr.')
df2.price=df2.price.str.replace(",",".")

df2['price']=df2['price'].astype(float)
df2['price']=df2['price'].round()
df2['price']=df2['price'].astype(int)

df2.head(2)
```

```python
colunms=['article_no','gender','category','price','size','brand']
df3=df2[colunms]
df3.head(2)
```

```python
npr=df3.to_numpy()
npr[0,4]
```

```python
r,c=npr.shape
print(r,c)
```

```python
import numpy as np
np.zeros([1,c])
```

```python
import numpy as np
npe=np.zeros((1,c))
    
for i in range(r):
    if npr[i,4]=='50-74':
        for s in [44,50,56,62,68,74]:
            npr[i,4]=s
#             print(npr[i,:])
            npe=np.concatenate((npe,npr[i:i+1,:]), axis=0)
    
    if npr[i,4]=='68-104':
        for s in [68,74,80,86,92,98,104]:
            npr[i,4]=s
#             print(npr[i,:])
            npe=np.concatenate((npe,npr[i:i+1,:]), axis=0)
    
    if npr[i,4]=='92-140':
        for s in [92,98,104,110,116,122,128]:
            npr[i,4]=s
    #             print(npr[i,:])
            npe=np.concatenate((npe,npr[i:i+1,:]), axis=0)

    if npr[i,4]=='134-170':
        for s in [134,130,136,152,158,164,170]:
            
            npr[i,4]=s
    #             print(npr[i,:])
            npe=np.concatenate((npe,npr[i:i+1,:]), axis=0)



    
```
