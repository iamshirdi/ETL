```python
import pandas as pd
result=pd.read_csv("Image-Classification-and-Price-prediction/scrapping/data/babyshop.csv", encoding='utf-8',index_col=False)
print(result.count())
result.head(3)
```

```python
df=result.drop_duplicates()
result.drop_duplicates().count()
```

```python
df['category'].unique()
```

```python
df['price'].unique()
```

```python
pd.set_option('display.max_colwidth', -1) #full colunm dispaly
df2=df
df2['price']=df2['price'].str.strip()
df2=df2[df2['price']!='- Få i lager']
df2=df2[df2['price']!='- En i lager']
df2=df2[df2['price']!='- I lager']
df2=df2[df2['price']!='$product.Price.Formatted']
df2=df2[df2['price']!='(Slut)']
df2=df2[df2['price']!= 'nan']



```

```python
df2['price'].unique()[10]
```

```python
df2['price'].unique()
```

```python
df2=df2.dropna() #na drop values
```

```python
pd.options.display.float_format = '{:,.0f}'.format

df2.price=df2.price.str.replace("\r", "")
df2.price=df2.price.str.strip(' kr')



df2['price']=df2['price'].astype(float)
df2['price']=df2['price'].round()
df2['price']=df2['price'].astype(int)

df2.head(2)
```

```python
df3=df2
df3['brand']=df3['brand'].str.strip()
df3.head(2)
```

```python
df3['size'].unique()
```

```python
sz_list=df3['size'].tolist()

new_sz=[]
for s in sz_list:
    
    if s.find("(")!=-1:
        s=s.split("(")[0]
#     s=s.replace(" ","")
    
    if s.find("years")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.replace('years','Y')
    
    
    if s.find("Years")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.replace('Years','Y')

    if s.find("months")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.replace('months','M')

        
    if s.find("år")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.replace('år','Y')

    if s.find("mån")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.replace('mån','M')
    
    if s.find("Months")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.replace('Months','M')
    
    
    if s.find("Month")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.replace('Month','M')
    
    
    if s.find("year")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.replace('year','Y')
    
    
    

    if s.find("m")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.replace('m','M')
    
    s=s.replace("cm","")
    s=s.replace("cM","")

    s=s.replace("\r","")
    s=s.replace("EU","")

    s=s.replace("+Y","Y")
    s=s.replace("y","Y")
    
    s=s.split(".")[0]
    s=s.split("/")[0]
    if s.find("-")!=-1:
        s=s.split("-")[1]
    
    s=s.replace(" Y","Y")
    s=s.replace(" M","M")
    s=s.replace("+","")
    
    if s.find(" ")!=-1:
        s=s.split(" ")[0]
    
    new_sz.append(s)
set(new_sz) #unique values
```

```python
new_siz=[]
for s in new_sz:
    if s=='One':
        s=1
        
    if s=='SMall':
        s='S'
    if s=='MediuM':
        s='M'
    if s=='Large':
        s='L'
    if s in ['T1','T2','Petite','Newborn','C10', 'C11', 'C12', 'C13', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'EL', 'J1', 'J2', 'J3','7GR'
            ,'37GR']:
        new_siz.append('Nan')
    else:
        new_siz.append(s)
set(new_siz)
```

```python
df3['new_size']=new_siz
```

```python
df3=df3[df3['new_size']!='Nan']
```

```python
df3['new_size'].unique()
```

```python
colunms=['article_no','gender','category','price','new_size','brand']
df4=df3[colunms]
df4.head(2)
```

```python
dfr=df4
dfr.to_csv('babyshop_cleaned.csv',index=False) #for rows not showing
```
