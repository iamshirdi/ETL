```python
pd.options.display.float_format = '{:,.0f}'.format
df2=df
df2.price=df2.price.str.strip(':-')

df2.price=df2.price.str.replace("\xa0", "")
df2.price=df2.price.str.replace(":",".")
df2.price=df2.price.str.replace(" ","")



df2['price']=df2['price'].astype(float)
df2['price']=df2['price'].round()
df2['price']=df2['price'].astype(int)

df2.head(2)
```

```python
df2['price'].unique()
```

```python
df2['size'].unique()
```

```python
sz_list=df2['size'].tolist()
```

```python
df2[df2['size']=='14-15 years'].count()
```

```python
new_sz=[]
for s in sz_list:
    
    s=s.rstrip('CM')
    s=s.rstrip('C')
    s=s.rstrip('CL')
    if s.find("/")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.split("/")[0]

        
    if s.find('.')!=-1:
        s=s.split('.')[0]
    
    if s.find(' years')!=-1:
        s=s.replace(' years','Y')
        
    
    if s.find(' months')!=-1:
        s=s.replace(' months','M')
     
    if s.find(' month')!=-1:
        s=s.replace(' month','M')
     
    if s.find(' year')!=-1:
        s=s.replace(' year','Y')

    if s.find('6-7YEARS')!=-1:
        s=s.replace('YEARS','Y')
    
    if s.find('6-7Y(S)')!=-1:
        s=s.replace('(S)','')
    
    if s.find('XX')!=-1:
        s=s.replace('XX','XXL')
    
    if s.find('ONE SIZE')!=-1:
        s=s.replace('ONE SIZE','1')
    
     
     
        
    new_sz.append(s)
set(new_sz) #unique values
```

```python
df2['new_size']=new_sz
```

```python
colunms=['article_no','gender','category','price','new_size','brand']
df3=df2[colunms]
df3.head(2)
```
