```python
import pandas as pd
pd.set_option('display.max_colwidth', -1) #full colunm dispaly

result=pd.read_csv("Image-Classification-and-Price-prediction/scrapping/data/zalando.csv", encoding='utf-8',index_col=False)
print(result.count())
result.head(3)
```

```python
df=result.drop_duplicates()
result.drop_duplicates().count()
```

```python
df['price'].unique()
```

```python
df[df['price']=='919\xa0kr']
```

```python
df2=df
df2['price']=df2['price'].str.strip()
df2['price']=df2['price'].str.replace('\xa0','')
df2['price']=df2['price'].str.replace('kr','')
df2['price']=df2['price'].str.replace(',','.')
df2['price']=df2['price'].astype(float)
df2['price']=df2['price'].round()
df2['price']=df2['price'].astype(int)


df2['price'].unique()
```

```python
df3=df2
df3['size']=df3['size'].str.replace('xREG','')

df3['size'].unique()
# df3[df3['size']=='56-80m']
```

```python
c=0
sz_list=df3['size'].tolist()
new_sz=[]
for s in sz_list:
    if s.find("/")!=-1:
        s=s.replace("/","-")
        
    if s in ['128xregular', '134xregular', '134xwide',
       '140xregular', '140xslim', '140xwide', '146xregular', '146xwide',
       '152xregular', '152xwide', '158xregular', '164xregular',
       '164xslim', '164xwide', '170xregular', '176xregular', '176xwide',
       '158xwide', '170xslim', '170xwide', '176xslim', '92xregular',
       '146xSLIM', '158xSLIM', '170xSLIM', '176xSLIM', '134xslim',
       '116xSLIM', '140xSLIM', '128xSLIM','60AA',
       '65AA', '65A', '70AA', '70A', '70B', '75A', '160-165', '60xAA',
       '65xA', '65xAA', '70xA', '70xAA', '75xA', '75xAA', '55A', '60A',
       '65B', '75AA', '116/122xSLIM','128/134xSLIM','134/140xSLIM','146/152xSLIM','158/164xSLIM','170/176xSLIM','92/98xSLIM', '56-80m',
             '104-110xSLIM']:
        c=c+1
        s='Na'
    if s.find("SLIM")!=-1:
        s='Na'
    s=s.replace("y","Y")
    if s.find(" ")!=-1:
        s=s.split(" ")[0]
    s=s.replace("One","One Size")
    new_sz.append(s)
    
print(c)
set(new_sz)
```

```python
df3['new_size']=new_sz
colunms=['article_no','gender','category','price','new_size','brand']
df4=df3[colunms]
df4.head(2)
```

```python
dfr=df4
dfr.to_csv('zalando_cleaned.csv',index=False) #for rows not showing
```
