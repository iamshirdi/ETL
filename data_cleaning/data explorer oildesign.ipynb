```python
sz_list=df3['size'].tolist()

new_sz=[]
for s in sz_list:
    if s.find("\n")!=-1:
        s=s.split("\n")[0]
    if s.find("cm")!=-1:
        s=s.replace("cm","")
    if s.find(" UK")!=-1:
        s='Na'
    if s.find("(")!=-1:
        s=s.split("(")[1]
        s=s.strip(")")
    
    if s.find("m")!=-1:
        s=s.split("m")[0]+'M'
    
    if s.find("M")!=-1:
        s=s.split("M")[0]+'M'
        
    if s.find("y")!=-1:
        s=s.replace("y","Y")
    
    if s.find("å")!=-1:
        s=s.split("å")[0]+'Y'  
    if s.find("Y")!=-1:
        s=s.split("Y")[0]+'Y'  
                
        
    s=s.replace("preM","premature")

    if s.find("/")!=-1:
        s=s.split("/")[1]
    s=s.replace(" ","")
    s=s.strip("+")
    s=s.replace("ÅR","Y")
    if s.find("+")!=-1:
        s='Na'
    if s.find("UK")!=-1:
        s='Na'
    if s.find("EU")!=-1:
        s='Na'
    if s.find("US")!=-1:
        s='Na'
 
    if s in ['$product.Name','$product.NaM','2XL160','BABY', 'BabY', 'DeniM','KY','Svart','S120', 'Strl0','Väljstorlek',
            '4мY','NY','L', 'L140','Väljstorlek','XXS100','XS110','XL150', '2XL160',
            'L140','PreM','S120', 'Strl0','Väljstorlek','XL150','XS110','XXS100']:
        s='Na'
    s=s.replace("1½Y","1.5")
    s=s.replace(",",".")
    new_sz.append(s)
set(new_sz)
```

```python
df3['new_size']=new_sz
df4=df3[df3['new_size']!='Na']

colunms=['article_no','gender','category','price','new_size','brand']
df5=df4[colunms]
df5.head(2)
```
