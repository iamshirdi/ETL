```python
new_sz=[]
for s in sz_list:
    
    s=s.strip('år')
    s=s.strip('månader')
    if s.find("/")!=-1:
#         print('splitting',s.split("/")[0])
        s=s.split("/")[0]
        
    if s.find('(')==-1:
        new_sz.append(s)
        continue
    start=s.find('(')
    end=s.find(')')
    new_sz.append(s[start+1:end].strip('cm'))
set(new_sz) #unique values
```

```python
df2['new_size']=new_sz
```
