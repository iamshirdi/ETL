```python
import pandas as pd
result=pd.read_csv("Image-Classification-and-Price-prediction/scrapping/data/boozt.csv", encoding='utf-8',index_col=False)
print(result.count())
result.head(3)
```

```python
df=result.drop_duplicates()
result.drop_duplicates().count()
```

```python
df['price'].unique()
```

```python
df2=df
# df2['price']=df2['price'].str.strip('')

df2['price']=df2['price'].astype(float)
df2['price']=df2['price'].round()
df2['price']=df2['price'].astype(int)

df2.head(2)
```

```python
df3=df2
df3['size']=df3['size'].str.strip()
df3['size']=df3['size'].str.strip('CM')

df3['size'].unique()
```

```python
sz_list=df3['size'].tolist()

new_siz=[]

for s in sz_list:
    
        
    if s.find("/")!=-1:
        s=s.split("/")[0]

    if s.find("-")!=-1:
        s=s.split("-")[1]
        
    if s.find("\\")!=-1:
        s=s.split("\\")[1]
    
    if s.find("MONTH")!=-1:
        s=s.replace("MONTH","M")
        
    s=s.strip('EU')
    if s.find("0 y")!=-1:
         s=s.replace("0 y","0Y")
        
    if s.find("01")!=-1:
        s=s.replace("01","1")
    
    if s.find("ONE SIZ")!=-1:
        s=s.replace("ONE SIZ","ONE SIZE")
    
    if s in ['100X100', '100X140', '100X70', '140X200', '140X220','16X12X27','30X21','30X360', '13.5W',
             '4A','6A', 

            '16X12X27','18X3.5','200ML','21X2.5','250ML','30X21', '30X360','40X40','50X63','70X100', '70X140','75X39','80X33', 
             '80X70','86\\92','90X90','ADULT', 'BABY', 'HILDREN','ONESIZE YOUTH','S1', 'S2', 'S3', 'VAL','YOUTH']:
        
        s='Na'
    if len(s)>1:
            s=s.strip("L")
    
    if s.find(".")!=-1:
        s=s.split(".")[0]
    if s.find(" ")!=-1:
        s=s.replace(" ","")
        
    if s in ['362', '371', '382', '391', '351', '552', '402', '411', '422', '431', '442', '451','471', '462']:
        s='Na'
        
    new_siz.append(s)
set(new_siz)
```

```python
df3['new_size']=new_siz
# df3['new_size']=df3[df3['new_size']!='Na']
```
