```python
import pandas as pd
pd.set_option('display.max_colwidth', -1) #full colunm dispaly

result=pd.read_csv("Image-Classification-and-Price-prediction/scrapping/data/lekmer.csv", encoding='utf-8',index_col=False)
print(result.count())
result.head(3)
```

```python
df=result.drop_duplicates()
result.drop_duplicates().count()
```

```python
df['price'].unique()
```

```python
df2=df
df2['price']=df2['price'].str.replace(' kr','')
df2['price']=df2['price'].astype(float)
df2['price']=df2['price'].round()
df2['price']=df2['price'].astype(int)

df2['price'].unique()
```

```python
df3=df2
df3['size']=df3['size'].str.strip()
df3['size']=df3['size'].str.replace(' ','')
df3['size']=df3['size'].str.replace('år','Y')
df3['size']=df3['size'].str.replace('years','Y')
df3['size']=df3['size'].str.replace('mån','M')
df3['size']=df3['size'].str.replace('Years','Y')
df3['size']=df3['size'].str.replace('months','M')
df3['size']=df3['size'].str.replace('Months','M')

df3['size']=df3['size'].str.replace('month','M')
df3['size']=df3['size'].str.replace('Mader','M')
df3['size']=df3['size'].str.split('(').str.get(0)
df3['size']=df3['size'].str.split('\n').str.get(0)

df3['size']=df3['size'].str.replace('cm','')


# df3['size'].unique()
```

```python
sz_list=df3['size'].tolist()

new_sz=[]
for s in sz_list:
    if s.find("m")!=-1:
        s=s.split('m')[0]+'M'
    if s.find("year")!=-1:
        s=s.replace("year","Y")
    if s.find("yrs")!=-1:
        s=s.replace("yrs","Y")
    if s.find("y")!=-1:
        s=s.replace("y","Y")
    if s in ['4/6-8Y','4Y/104','5Y/110','6-12M/20-21','6Y/116','9Y/134', '3Y/98', '0-6M/16-19']:
        s=s.replace("4/6-8Y","6-8Y")
        s=s.replace('3Y/98',"3Y")
        s=s.replace('0-6M/16-19',"0-6M")

        s=s.replace("4Y/104","4Y")
        s=s.replace("5Y/110","5Y")
        s=s.replace("6-12M/20-21","6-12M")
        s=s.replace("6Y/116","6Y")
        s=s.replace("9Y/134","9Y")
    s=s.replace("'BabY0-3Y'","0-3Y")      
    s=s.replace('24-36M/24-25',"24-36M") 
    
    s=s.replace('OneSize1-3Y',"One Size")
    s=s.replace('PreM',"Premature")
    s=s.replace('CM',"")
    
    
    if s in ['BABY', 'BabY' 'Gul', 'J1', 'J2', 'J3', 'JR','M2/W4', 'Mini', 'NB', 'NYfödd','T1','YMD','YXS',
            'BabY', 'BabY0-3Y', 'Blå', 'Gul','Newborn', 'ONESIZE', 'One Size', 'OneSize', 'Premature','Strl1', 'Svart',
             'UK13-2', 'UK3-5', 'US11', 'US13', 'US4',
            'C10', 'C13', 'C2-3', 'C4', 'C4/5', 'C4/C5', 'C5', 'C6', 'C7',]:
        s='Na'

    s=s.replace(',',"-")
    s=s.replace('/',"-")

    new_sz.append(s)
set(new_sz)
```
