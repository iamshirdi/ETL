```python
import pandas as pd
pd.set_option('display.max_colwidth', -1) #full colunm dispaly

result=pd.read_csv("Image-Classification-and-Price-prediction/scrapping/data/polarnopyret.csv", encoding='utf-8',index_col=False)
print(result.count())
result.head(3)
```

```python
df=result.drop_duplicates()
result.drop_duplicates().count()
```

```python
df['price'].unique()
```

```python
df2=df
df2['price']=df2['price'].str.split(' ').str.get(0)

df2['price'].unique()
```

```python
df3=df2
df3['size'].unique()
```

```python
df3[df3['size']=='1 3']
```

```python
df3['size']=df3['size'].str.replace('one size','One size')
df3['size']=df3['size'].str.replace('Onesize','One size')
df3['size']=df3['size'].str.replace(' ','-')
df3['size'].unique()
```

```python
colunms=['article_no','gender','category','price','size','brand']
df4=df3[colunms]
df4.head(2)
```
